```python
# new new
class LanguageModelDataset(torch.utils.data.Dataset):
    def __init__(self, tokens, token2index, pad_index, edge_index, max_len=50):
        self.tokens = tokens
        self.token2index = token2index
        self.pad_index = pad_index
        self.edge_index = edge_index
        self.max_len = max_len

    def __len__(self):
        return len(self.tokens)

    def __getitem__(self, index):
        # Get the current word (as a string)
        token = self.tokens[index]
    
        # Convert the word into a sequence of character indices, adding edge tokens
        token_indices = [self.edge_index] + [
            self.token2index.get(char, self.token2index['<UNK>']) for char in token
        ] + [self.edge_index]
    
        # Truncate or pad the sequence
        if len(token_indices) > self.max_len:
            token_indices = token_indices[:self.max_len]
        else:
            token_indices += [self.pad_index] * (self.max_len - len(token_indices))
    
        # Convert to tensor
        token_tensor = torch.tensor(token_indices, dtype=torch.long)
    
        # Inputs and targets (for next-character prediction)
        inputs = token_tensor[:-1]
        targets = token_tensor[1:]
    
        return inputs, targets

```

```python
# Defining the GRU Language Model (new)
class GRUModel(torch.nn.Module):

    def __init__(self, vocab_size, embedding_size, state_size, pad_index, dropout_rate, init_stddev):
        super().__init__()
        self.pad_index = pad_index

        self.embedding_layer = torch.nn.Embedding(vocab_size, embedding_size)
        self.rnn_c0 = torch.nn.Parameter(torch.zeros((state_size,), dtype=torch.float32))
        self.rnn_h0 = torch.nn.Parameter(torch.zeros((state_size,), dtype=torch.float32))
        self.rnn_cell = torch.nn.GRUCell(embedding_size, state_size)
        self.dropout = torch.nn.Dropout(dropout_rate)
        self.output_layer = torch.nn.Linear(state_size, vocab_size)
        self.init_stddev = init_stddev

    def initialise(self, seed):
        g = torch.Generator()
        g.manual_seed(seed)
        for (name, param) in self.named_parameters():
            torch.nn.init.normal_(param, std=self.init_stddev, generator=g)

    def forward(self, text_indexes):
        (batch_size, num_time_steps) = text_indexes.shape
    
        pad_mask = text_indexes == self.pad_index
        embedded = self.embedding_layer(text_indexes)
        state = self.rnn_h0[None, :].repeat(batch_size, 1)
        state = self.dropout(state)
    
        outputs = []
        for t in range(num_time_steps):
            new_state = self.rnn_cell(embedded[:, t, :], state)
            state = torch.where(pad_mask[:, t, None], state, new_state)
            outputs.append(state)
    
        # Stack all outputs and apply the output layer
        outputs = torch.stack(outputs, dim=1)  # Shape: [batch_size, seq_len, state_size]
        logits = self.output_layer(outputs)  # Shape: [batch_size, seq_len, vocab_size]
        return logits
```

```python
def process_texts_set(set_tokens, token2index, pad_index, max_len):
    set_x = []
    set_t = []

    for word in set_tokens:
        # Convert word to indexed characters with <EDGE> tokens
        indexed_word = [token2index['<EDGE>']] + \
                       [token2index.get(char, token2index['<UNK>']) for char in word] + \
                       [token2index['<EDGE>']]
        
        # Truncate or pad indexed_word to max_len
        if len(indexed_word) > max_len:
            indexed_word = indexed_word[:max_len]
        else:
            indexed_word += [pad_index] * (max_len - len(indexed_word))
        
        # Inputs (excluding last token) and targets (excluding first token)
        set_x.append(indexed_word[:-1])
        set_t.append(indexed_word[1:])

    # Convert to tensors
    set_x = torch.tensor(set_x, dtype=torch.long)
    set_t = torch.tensor(set_t, dtype=torch.long)

    return set_x, set_t
```

## 3) Generation and evaluation

Now you need to use your language model with a beam search algorithm in order to expand partially written words into the most probable full words.
It's important to be able to test the beam search algorithm once it is written, but since the predictions of the language model are not predictable and thus a test code using it is also unpredictable, we'll need to use a mock model.
A mock model is a substitute for the trained model but which does not use any machine learning, only hard coded values.

The below mock model works just like a language model but gives a high probability to the following words only: a, the, these, and alphabet.
Any input prefix which is not a prefix of one of the previous four words will encourage a beam search algorithm to stop generating anything further.

```python
class Mock(torch.nn.Module):

    def __init__(self, vocabulary, token2index, pad_index, edge_index):
        super().__init__()
        self.vocabulary = vocabulary
        self.token2index = token2index
        self.pad_index = pad_index
        self.edge_index = edge_index

    def _mock_prediction(self, prefix):
        logits = torch.zeros((len(vocabulary),), dtype=torch.float32, device='cpu')
        predictions = {
            '': ['t', 'a'],
            't': ['h'],
            'th': ['e'],
            'the': ['\0', 's'],
            'thes': ['e'],
            'these': ['\0'],
            'a': ['\0', 'l'],
            'al': ['p'],
            'alp': ['h'],
            'alph': ['a'],
            'alpha': ['b'],
            'alphab': ['e'],
            'alphabe': ['t'],
            'alphabet': ['\0'],
        }.get(prefix, ['\1'])
        for p in predictions:
            if p == '\0':
                logits[self.edge_index] = 10.0
            elif p == '\1':
                logits[self.edge_index] = 0.01
            else:
                logits[self.token2index[p]] = 10.0
        return logits

    def forward(self, text_indexes):
        prefixes = [
            ''.join(self.vocabulary[index] for index in indexes if index not in [pad_index, edge_index])
            for indexes in text_indexes.cpu().tolist()
        ]
        logits = torch.stack([self._mock_prediction(prefix) for prefix in prefixes], dim=0)
        logits = logits[:, None, :].tile((1, text_indexes.shape[1], 1))
        return logits


mock_model = Mock(vocabulary, token2index, pad_index, edge_index)
```

```python
# new
def beam_generate(model, vocabulary, token2index, max_len=10, beam_size=3, prefix=''):
    unk_index = token2index['<UNK>']
    edge_index = token2index['<EDGE>']

    # Convert the prefix into token indexes
    prefix_tokens = list(prefix)  # Convert prefix string into a list of characters
    beam_prefix_indexes = torch.tensor([
        [edge_index] + [token2index.get(token, unk_index) for token in prefix_tokens]
    ], dtype=torch.int64, device='cpu')
    beam_prefix_probs = np.array([1.0], np.float32)

    # Best full word tracking
    best_full_word = None
    best_full_word_prob = 0.0

    with torch.no_grad():
        for _ in range(max_len):
            # Get predictions from the model
            logits = model(beam_prefix_indexes)
            probs = torch.softmax(logits, dim=2)
            token_probs = probs[:, -1, :].cpu().numpy()
            token_probs[:, unk_index] = 0.0  # Ignore <UNK>

            # Expand the beam
            new_beam = []
            for i, prefix in enumerate(beam_prefix_indexes):
                for token_index, prob in enumerate(token_probs[i]):
                    if prob > 0:  # Only consider tokens with non-zero probability
                        new_prefix = prefix.tolist() + [token_index]
                        new_prob = beam_prefix_probs[i] * prob
                        new_beam.append((new_prefix, new_prob))

            # Sort by probability and prune the beam
            new_beam.sort(key=lambda x: x[1], reverse=True)
            new_beam = new_beam[:beam_size]

            # Update beam data structures
            beam_prefix_indexes = torch.tensor(
                [prefix for prefix, _ in new_beam], dtype=torch.int64, device='cpu'
            )
            beam_prefix_probs = np.array([prob for _, prob in new_beam], dtype=np.float32)

            # Check for completed words
            for prefix, prob in new_beam:
                if prefix[-1] == edge_index:  # If the sequence ends with <EDGE>
                    full_word = ''.join([vocabulary[token] for token in prefix[1:-1]])
                    if prob > best_full_word_prob:
                        best_full_word = full_word
                        best_full_word_prob = prob

            # If the beam only contains completed words, stop early
            if all(prefix[-1] == edge_index for prefix, _ in new_beam):
                break

    return best_full_word or prefix  # If no valid word is found, return the prefix
```

3.3) Calculate and display the percentage of taps saved when using the text predictor instead of typing all the words in the test set.
This is calculated as

$$\text{taps\_saved}(\text{words}) = 1 - \frac{\sum_{w \in \text{words}} \text{taps\_needed}(w)}{\sum_{w \in \text{words}} \text{len}(w)}$$

Make sure that you use the real language model you trained (not the mock model) and the words in the test set (as strings).

[5 marks]

```python
def calculate_taps_saved(model, words, vocabulary, token2index, max_len=10, beam_size=3):
    total_taps_needed = 0
    total_taps_typing = 0

    for word in words:
        # Taps needed using the predictor
        taps = taps_prediction(model, word, vocabulary, token2index, max_len, beam_size)
        total_taps_needed += taps

        # Taps needed without the predictor (length of the word)
        total_taps_typing += len(word)

    # Calculate taps saved percentage
    taps_saved = 1 - (total_taps_needed / total_taps_typing)
    return taps_saved * 100
```
